```python
import kivy
kivy.require('1.11.1')
from kivy.app import App
from kivy.uix.button import Button
from kivy.uix.gridlayout import GridLayout
from kivy.clock import Clock
from kivy.core.window import Window

import os
import ftplib
import glob

class MyApp(App):

    def build(self):
        # Create the main layout
        layout = GridLayout(cols=1, spacing=10)

        # Create the button
        self.send_button = Button(text='Send Images')
        self.send_button.bind(on_press=self.send_images)

        # Add the button to the layout
        layout.add_widget(self.send_button)

        # Create the FTP object
        self.ftp = ftplib.FTP('ftp.example.com', 'username', 'password')

        # Keep track of the sending status
        self.sending = False

        # Schedule the sending function to run every minute
        Clock.schedule_interval(self.send_images, 60)

        return layout

    def send_images(self, instance):
        if self.sending:
            return

        # Enable/disable the appropriate button
        self.send_button.disabled = True

        # Set the sending status
        self.sending = True

        # Get a list of the most recent images
        image_files = sorted(glob.glob(os.path.join(Window.storage.get('pictures'), '*.jpg')), key=os.path.getmtime, reverse=True)[:10]

        # Upload each image to the FTP server
        for image_file in image_files:
            with open(image_file, 'rb') as f:
                self.ftp.storbinary('STOR ' + os.path.basename(image_file), f)

        # Reset the sending status and enable the button
        self.sending = False
        self.send_button.disabled = False
```
